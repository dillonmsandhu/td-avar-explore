```python
import sys
from pathlib import Path

# notebooks → deep
PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

print("PROJECT_ROOT:", PROJECT_ROOT)

import numpy as np
import pandas as pd

from utils import load_run_data

```

```python
import numpy as np
import pandas as pd
RESULTS_ROOT = PROJECT_ROOT / "results"

RESULTS_LOOKUP = {
    "cov_lstd_optimistic": "cov_lstd_opt",
    "cov_net_v": "cov_net_v",
}

Ns = range(30, 51)
ENV_NAME = "DeepSea-bsuite"
DATE = "2026-01-14"

results = {}

for file, subdir in RESULTS_LOOKUP.items():
    results[file] = {}

    for N in Ns:
        run_name = f"{subdir}/deepsea_sweep/{DATE}/{file}_N{N}_E_false"

        try:
            config, out = load_run_data(
                run_folder_name=run_name,
                env_name=ENV_NAME,
                results_base_path=str(RESULTS_ROOT),
            )

            rets = out["returned_episode_returns"]
            per_seed_last_ten_steps_mean_ret = out["returned_episode_returns"][:, -20:].mean(-1)
            seeds_that_solve = np.sum(per_seed_last_ten_steps_mean_ret > 0.25)
            # seeds_that_win = np.sum(out["returned_episode_returns"].mean( > 0.5)

            results[file][N] = float(seeds_that_solve)

        except FileNotFoundError:
            print(RESULTS_ROOT / run_name / ENV_NAME)
            results[file][N] = np.nan

DATE = "2026-01-15"
RESULTS_LOOKUP = {
    "cov_lstd_opt_simple": "cov_lstd_opt_simple",
}

for file, subdir in RESULTS_LOOKUP.items():
    results[file] = {}

    for N in Ns:
        run_name = f"{subdir}/deepsea_sweep/{DATE}/{file}_N{N}_E_false_N0_1000"

        try:
            config, out = load_run_data(
                run_folder_name=run_name,
                env_name=ENV_NAME,
                results_base_path=str(RESULTS_ROOT),
            )

            rets = out["returned_episode_returns"]
            per_seed_last_ten_steps_mean_ret = out["returned_episode_returns"][:, -20:].mean(-1)
            seeds_that_solve = np.sum(per_seed_last_ten_steps_mean_ret > 0.25)
            # seeds_that_win = np.sum(out["returned_episode_returns"].mean( > 0.5)

            results[file][N] = float(seeds_that_solve)

        except FileNotFoundError:
            print(RESULTS_ROOT / run_name / ENV_NAME)
            results[file][N] = np.nan
```
